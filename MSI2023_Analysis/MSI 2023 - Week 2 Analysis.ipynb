```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patheffects as path_effects
import seaborn as sns

import datetime
from PIL import Image as pImg

pd.options.display.max_columns = 100
```

### About the dataset

The datset contains all the proffesional League of Legends games played in a tournament setting in 2023.
It has been aggregated and released by Tim Sevenhuysen of OraclesElixir.com, can be found at https://oracleselixir.com/tools/downloads

```python
base_df = pd.read_csv("ESPORTS_DATA.csv")
base_df = base_df.query('datacompleteness == "complete"')
base_df.date = pd.to_datetime(base_df.date).dt.date
```

### Remove unnecessary columns

```python
base_df = base_df.drop(columns=['gameid','datacompleteness','url','split','year','playoffs','patch','participantid','playerid','teamid','opp_elementaldrakes','opp_elders','opp_heralds','opp_barons','opp_towers','opp_turretplates','opp_inhibitors','opp_killsat10','opp_assistsat10','opp_deathsat10','opp_goldat15','opp_xpat15','opp_csat15','opp_killsat15','opp_assistsat15','opp_deathsat15'])
```

### Filter MSI 2023 data

```python
msi_start_date = datetime.date(2023, 5, 2)
msi_df = base_df.query('league == "MSI" & date >= @msi_start_date')
msi_df_teams = msi_df.query('position == "team"')
msi_df_players = msi_df.query('position != "team"')
```

### Split by play-ins/group stage

```python
last_playin_day = datetime.date(2023, 5, 7)
last_groupstage_week1_day = datetime.date(2023, 5, 14)
py_df = msi_df.query('date <= @last_playin_day')
gs_df = msi_df.query('date > @last_playin_day & date <= @last_groupstage_week1_day')
```

### Split by team / players

```python
py_df_teams = py_df.query('position == "team"')
py_df_players = py_df.query('position != "team"')
gs_df_teams = gs_df.query('position == "team"')
gs_df_players = gs_df.query('position != "team"')
```

### Data accesibility functions

```python
def get_match_no(game_type = 0):
    """
    Returns the total number of games played for a given game type.

    Args:
         game_type (int): The game type to filter by. 
            0: Group stage
            1: Play-ins stage
            2: Combined (both group stage and play-ins)

    Returns:
        float: The total number of games played for the given game type.
    """
    if game_type == 0:
        return gs_df_teams.shape[0] / 2
    elif game_type == 1:
        return py_df_teams.shape[0] / 2
    elif game_type == 2:
        return gs_df_teams.shape[0] / 2 + py_df_teams.shape[0] / 2
    else:
        raise ValueError('Invalid game type')

def champion_picks(game_type = 0):
    """Returns the number of matches for all champion that were picked at least once for a given game type.

    Args:
        game_type (int): The game type to filter by. 
            0: Group stage
            1: Play-ins stage
            2: Combined (both group stage and play-ins)

    Returns:
        Series: A series containing all the champions that were picked at least once and the number of picks.
    """
    if game_type == 0:
        return gs_df_players.groupby('champion').count().sort_values(by=['result'], ascending = False)['result']
    elif game_type == 1:
        return py_df_players.groupby('champion').count().sort_values(by=['result'], ascending = False)['result']
    elif game_type == 2:
        return msi_df_players.groupby('champion').count().sort_values(by=['result'], ascending = False)['result']
    else:
        raise ValueError('Invalid game type')
        
def top_pick_by_role(role = None, game_type = 0):
    if role != None:
        if game_type == 0:
            filtered_by_role = gs_df_players[gs_df_players.position == role]
        elif game_type == 1:
            filtered_by_role = py_df_players[py_df_players.position == role]
        elif game_type == 2:
            filtered_by_role = msi_df_players[msi_df_players.position == role]
        else:
            raise ValueError('Invalid game type')
        top_picked_df = filtered_by_role[['champion', 'result']].groupby('champion').count().sort_values(by=['result'], ascending = False)
        most_picked = top_picked_df.squeeze().head().sort_values()
    return most_picked
        
def champion_bans(game_type = 0):
    """Returns the number of matches a champion was banned for all champions that were banned at least once a given game type.

    Args:
        game_type (int): The game type to filter by. 
            0: Group stage
            1: Play-ins stage
            2: Combined (both group stage and play-ins)

    Returns:
        Series: A series containg all the champions that were banned at least once and the number of bans for a given game type.
    """
    if game_type == 0:
        filtered_df = gs_df_teams
    elif game_type == 1:
        filtered_df = py_df_teams
    elif game_type == 2:
        filtered_df = msi_df_teams
    else:
        raise ValueError('Invalid game type')
    return pd.concat([filtered_df['ban1'], filtered_df['ban2'], filtered_df['ban3'], filtered_df['ban4'], filtered_df['ban5']]) \
         .value_counts() \
         .rename_axis('champion')
    
        
def champion_match_no(champ_name, game_type = 0):
    """Returns the number of matches a champion was played in for a given game type.

    Args:
        champ_name (str): The name of the champion to look for.
        game_type (int): The game type to filter by. 
            0: Group stage
            1: Play-ins stage
            2: Combined (both group stage and play-ins)

    Returns:
        int: The number of matches the champion was played in for the given game type.
    """
    return champion_picks(game_type).loc[new_capitalize(champ_name)]

def ban_no(champ_name, game_type = 0):
    """
    Returns the number of bans a champion had in a given game type.

    Parameters:
    champ_name (str): Name of the champion to get the number of bans for.
    game_type (int): The game type to filter by. 
            0: Group stage
            1: Play-ins stage
            2: Combined (both group stage and play-ins)

    Returns:
    int: The number of bans the champion had.
    """
    return champion_bans(game_type)[new_capitalize(champ_name)]
                        
def pick_rate(champ_name, game_type = 0):
    """
    Returns the pick rate for a champion in the given game type.

    Args:
        champ_name (str): Name of the champion to calculate the pick rate for.
        game_type (int): The game type to filter by. 
            0: Group stage
            1: Play-ins stage
            2: Combined (both group stage and play-ins)

    Returns:
        float: Pick rate of the champion rounded to two decimal places.

    """
    return round(champion_match_no(champ_name, game_type) / get_match_no(game_type), 2)

def ban_rate(champ_name, game_type = 0):
    """
    Returns the ban rate for a champion in the given game type.

    Args:
        champ_name (str): Name of the champion to calculate the ban rate for.
        game_type (int): The game type to filter by. 
            0: Group stage
            1: Play-ins stage
            2: Combined (both group stage and play-ins)

    Returns:
        float: Ban rate of the champion rounded to two decimal places.

    """
    return round(ban_no(champ_name, game_type) / get_match_no(game_type), 2)

def presence_rate(champ_name, game_type = 0):
    """
    Returns the presence rate for a champion in the given game type.

    Args:
        champ_name (str): Name of the champion to calculate the presence rate for.
        game_type (int): The game type to filter by. 
            0: Group stage
            1: Play-ins stage
            2: Combined (both group stage and play-ins)

    Returns:
        float: Presence rate of the champion rounded to two decimal places.

    """
    return round(pick_rate(champ_name, game_type) + ban_rate(champ_name, game_type),2)
```

### Visualization functions

```python
def generate_bar_chart(data, title, xlabel=None, ylabel=None, ax = None, large_images = False, img_size_big = -4):
    bar_palette = red_palette = sns.color_palette(["#f94144", "#f3722c", "#f8961e", "#f9c74f", "#90be6d", "#43aa8b", "#577590"])
    sns.set_palette(bar_palette)
    sns.set_style('darkgrid')
    if ax is None:
        fig, ax = plt.subplots()
    bar = ax.bar(data.index, data.values, color=bar_palette)
    ticks = ax.get_xticks()
    ax.set_xticks(ticks)
    ax.set_xticklabels(ax.get_xticklabels(), fontweight='bold')
    ticks = ax.get_yticks()
    ax.set_yticks(ticks)
    ax.set_yticklabels(ax.get_yticklabels(), fontweight='bold')
    ax.set_title(title, fontname='DejaVu Sans', fontweight='bold')
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    
    for i, (label, value) in enumerate(data.items()):
        #Top text
        ax.text(i, value-1, str(value) + " games", ha='center', va='bottom', \
                color='w',\
                fontweight='bold',\
                path_effects=[path_effects.Stroke(linewidth=2.5, foreground='black'),
                      path_effects.Normal()])
        #Picture
        try:
            champ_img = pImg.open('ChampSquares/{}Square.png'.format(label))
            if large_images == False:
                ax.imshow(champ_img, extent=[i-0.265, i+0.265, 0.25, 2.65], aspect='auto', zorder=2)
                ax.set_ylim(0, max(data.values) + 1)
            else:
                ax.imshow(champ_img, extent=[i-0.385, i+0.395, img_size_big, 0], aspect='auto', zorder=2)
                ax.set_ylim(img_size_big, max(data.values) + 1)
            ax.set_xlim(-0.5, data.shape[0] - 0.5)
        except:
            pass
    return ax
```

```python
def new_capitalize(s):
    return "'".join(word.capitalize() for word in s.split("'"))
```

### Most impactful objectives

```python
first_tower_wr = gs_df_teams.query('firsttower == 1 & result == 1')['firstblood'].count() / get_match_no()
first_blood_wr = gs_df_teams.query('firstblood == 1 & result == 1')['firstblood'].count() / get_match_no()
first_herald_wr = gs_df_teams.query('firstherald == 1 & result == 1')['firstherald'].count() / get_match_no()
first_baron_wr = gs_df_teams.query('firstbaron == 1 & result == 1')['firstbaron'].count() / get_match_no()
elder_wr = gs_df_teams.query('elders == 1 & result == 1')['elders'].count() / gs_df_teams.query('elders == 1')['elders'].count()
first_threetowers_wr = gs_df_teams.query('firsttothreetowers == 1 & result == 1')['firsttothreetowers'].count() / get_match_no()

# Calculate win rates
win_rates = [first_tower_wr, first_blood_wr, first_herald_wr, first_baron_wr, elder_wr, first_threetowers_wr]
labels = ['First Tower', 'First Blood', 'First Herald', 'First Baron', 'Elder Dragon', 'First Three Towers']

obj_series = pd.Series(win_rates, labels).sort_values(ascending=False)
obj_series.index.name = 'objective'
obj_series = (obj_series * 100).astype(int)
sns.set_style('darkgrid')
bar_palette = sns.color_palette("coolwarm", len(obj_series))

# Create the plot
fig, ax = plt.subplots(figsize=(8, 6))
bar = sns.barplot(x=obj_series.values, y=obj_series.index, palette=bar_palette)

# Add labels and title
ax.set_xlabel('Win Rate (%)')
ax.set_ylabel('Objective')
ax.set_title('Win rate by objective')

# Format the x-axis ticks as percentages
ax.set_xticklabels(['{}%'.format(int(x)) for x in ax.get_xticks()])

# Add value labels on each bar
for i, v in enumerate(obj_series.values):
    ax.text(v + 0.5, i, str(v) + '%', va='center', fontweight='bold')

# Show the plot
plt.show()
```
